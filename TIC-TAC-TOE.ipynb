```python
from IPython.display import clear_output

def print_board(board):
    print(board[:3])
    print(board[3:6])
    print(board[6:])
```

```python
def user_input_XO():
    choice = 'WRONG'
    
    ##CHOICE is X or O in TIC-TAC-TOE
    while choice not in ['X','O','x','o']:
        
        ##ENTER CHOICE FOR PLAYER 1 
        choice = input("\nEnter the marker you would like (X or O) : ")
        if choice not in ['X','O','x','o']:
            clear_output()
            print('\nSorry! Please enter Either X or O')
    clear_output()
    
    ##CHOICE FOR PLAYER 2
    if choice in ['X','x']:
        other_choice = 'O'
    else:
        other_choice = 'X'
    return choice.upper(),other_choice
```

```python
def user_input(board,choices,player_choice):
    choice = 'WRONG'
    
    while choice not in choices:
        choice = input("\nEnter the square where marker has to be placed(1-9) : ")
        if choice not in ['1','2','3','4','5','6','7','8','9']:
            clear_output()
            print('\nSorry! Please enter a number between 1-9')
        else:
            print('\n{} already placed, Please reenter another choice'.format(choice))
    clear_output()
    choices.remove(choice)
    choice = int(choice)
    if player_choice == 'X':
        board[choice-1] = 'X'
    else:
        board[choice-1] = 'O'
    print_board(board)
```

```python
def winner(board,choices,player_choice,win):
    if board[0] == board[1] == board[2] == player_choice:
        print('\n{} is the winner'.format(player_choice))
        win = 1
        return win
    elif board[3] == board[4] == board[5] == player_choice:
        print('\n{} is the winner'.format(player_choice))
        win = 1
        return win
    elif board[6] == board[7] == board[8] == player_choice:
        print('\n{} is the winner'.format(player_choice))
        win = 1
        return win
    elif board[0] == board[3] == board[6] == player_choice:
        print('\n{} is the winner'.format(player_choice))
        win = 1
        return win
    elif board[1] == board[4] == board[7] == player_choice:
        print('\n{} is the winner'.format(player_choice))
        win = 1
        return win
    elif board[2] == board[5] == board[8] == player_choice:
        print('\n{} is the winner'.format(player_choice))
        win = 1
        return win
    elif board[2] == board[4] == board[6] == player_choice:
        print('\n{} is the winner'.format(player_choice))
        win = 1
        return win
    elif board[0] == board[4] == board[8] == player_choice:
        print('\n{} is the winner'.format(player_choice))
        win = 1
        return win
    else:
        if(len(choices) == 0):
            print('\nNo winner')
            win = 1
            return win
        else:
            win = 0
            return win
```

```python
def user_input_yes_no():
    choice = 'WRONG'
    
    
    while choice not in ['Y','N','y','n']:
         
        choice = input("\nWould you like to play again? Yes(Y) or No(N)")
        if choice not in ['Y','y','N','n']:
            clear_output()
            print('\nSorry! Please enter Either Y for Yes or N for No')
    clear_output()
    
    if choice in ['Y','y']:
        choice = 1
    else:
        choice = 0
    return choice
```

```python
play = 1

while play == 1:
    board = [' ',' ',' ',' ',' ',' ',' ',' ',' ']
    choices = ['1','2','3','4','5','6','7','8','9']
    win = 0
    print('Welcome to the Game of TIC-TAC-TOE')
    player1_choice,player2_choice = user_input_XO()
    print('\nPlayer 1 Choice is {} \nPlayer 2 Choice is {}'.format(player1_choice,player2_choice))
    print('\nInitial Board arrangement ')
    print_board(board)
    while win == 0:
        print('Player 1 Turn')
        user_input(board,choices,player1_choice)
        win = winner(board,choices,player1_choice,win)
        if win == 0:
            print('Player 2 Turn')
            user_input(board,choices,player2_choice)
            win = winner(board,choices,player2_choice,win)
    play = user_input_yes_no()
```
